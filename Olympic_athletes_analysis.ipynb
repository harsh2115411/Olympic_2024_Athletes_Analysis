```python
import pandas as pd
import matplotlib.pyplot as plt
import time,datetime
```

```python
df = pd.read_csv('AthletesCSV.csv')
```

```python
df.head()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df = df.drop(['current'],axis = 1)
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop(columns = ['nationality', 'country', 'weight', 'nationality_code', 'hero', 'occupation', 'reason', 'ritual', 'hobbies', 'residence_place'], inplace = True )
df.drop(columns = ['sporting_relatives', 'height', 'name_short', 'name_tv'], inplace =True)
df.drop(columns = ['philosophy', 'influence'], inplace =True)
df['disciplines']= df['disciplines'].str.strip("['']123._/")
df['events']= df['events'].str.strip("['']123._/")
df['events']= df['events'].str.strip('""')
df['name'] = df['name'].str.strip()
df['name']= df['name'].apply(lambda x: x.title())
df = df.fillna(' ')
df['birth_place']= df['birth_place'].apply(lambda x: x.title())
df['sport'] = df['disciplines']
```

```python
df.info()
```

```python
df.head()
```

```python
df.drop(columns = ['education','family','other_sports'],inplace = True)
```

```python
df.head()
```

```python

df['birth_date'] = pd.to_datetime(df['birth_date'])
```

```python
df.info()
```

```python
df['coach'] = df['coach'].str.replace('<br>', '')
```

```python
df[['sports', 'sport1']]=df['disciplines'].str.split(',', n=1, expand=True)
df[['sport2', 'sport3']]=df['sports'].str.split("'", n=1, expand=True)
```

```python
a= df['sport2'].value_counts()
b = df['sport1'].value_counts()
print(f'{a} {b}')
```

```python
a = athlete.plot(x= 'sport2', kind ='bar',  figsize=(12, 4))
a.set_ylabel('Number of Athletes')
a.set_xlabel('Type Of Sport')
a.set_title('Number Of Athletes by Sport Types')
plt.show()
```

```python
ath_country = df['country_long'].value_counts()
b = ath_country.plot(y= 'country_long', kind ='barh', figsize=(10, 40))
b.set_ylabel('Country')
b.set_xlabel('Number of Athletes')
b.set_title('Number of Athletes by Country')
plt.show()
```
